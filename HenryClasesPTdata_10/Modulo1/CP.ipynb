```python

dat = {'A': [1, 2, 3],
        'B': [4, 5, 6],
        'C': [7, 8, 9]}
df = pd.DataFrame(dat, index=['x', 'y', 'z'])
```

```python
df.iloc[2]["A"]
```

```python
# Usando loc para acceder a un dato por etiquetas de fila y columna
print(df.loc['y', 'B'])  # Salida: 5

# Usando iloc para acceder a un dato por posici√≥n entera de fila y columna
print(df.iloc[1, 1])  # Salida: 5
```

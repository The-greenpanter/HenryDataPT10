```python
Solucion punto 1 b

def decimal_a_binario(decimal):
    parte_entera = int(decimal)
    parte_fraccionaria = decimal - parte_entera

    binario_parte_entera = bin(parte_entera)[2:]

    binario_parte_fraccionaria = ""
    while parte_fraccionaria != 0:
        parte_fraccionaria *= 2
        bit = int(parte_fraccionaria)
        binario_parte_fraccionaria += str(bit)
        parte_fraccionaria -= bit

    return binario_parte_entera + "." + binario_parte_fraccionaria

fracciones = [1/2, 1/3, 1/4, 1/5, 1/6, 1/7, 1/8, 1/9]

for fraccion in fracciones:
    binario = decimal_a_binario(fraccion)
    print("Fraccion 1 / {} : {:.16f}  y En binario:  {}".format(int(1/fraccion), fraccion, binario))
```
